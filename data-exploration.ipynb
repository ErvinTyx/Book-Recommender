```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
df = pd.read_csv(f"{path}/books.csv")
```

```python
df
```

```python
def column_stats(df):
    total_rows = len(df)
    stats = {}
    for col in df.columns:
        col_data = df[col]
        missing = col_data.isna().sum()
        missing_pct = (missing / total_rows) * 100 if total_rows > 0 else 0
        
        distinct = col_data.nunique(dropna=True)
        count = col_data.count()  # non-missing count
        
        # Determine the top (most frequent) value and its frequency.
        mode_series = col_data.mode(dropna=True)
        if not mode_series.empty:
            top = mode_series.iloc[0]
            frequency = col_data.value_counts().iloc[0]
            frequency_pct = (frequency / count * 100) if count > 0 else 0
        else:
            top, frequency, frequency_pct = None, 0, 0

        stats[col] = {
            'Missing': missing,
            'Missing_Percentage': missing_pct,
            'Distinct': distinct,
            'Count': count,
            'Top': top,
            'Frequency': frequency,
            'Frequency_Percentage': frequency_pct
        }
        
    return pd.DataFrame(stats).T 
```

```python
column_stats(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(),cbar=False,ax = ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np

df['missing_description'] = np.where(df['description'].isna(),1,0)
df["age_of_book"] = 2025 - df["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book","missing_description","average_rating"]

# using correlation to dealing with completely continuos varaibles
# using spearman to deal with non-continuos values
correlation_matrix = df[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(
    correlation_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm",
    cbar_kws={"label":"Correlation"}
)
heatmap.set_title("Correlation heatmap")
plt.show()
```

# From the heat map above 
we know that missing desciption does not have much correated with other value then itself.
The reason why we want to know this say number of pages was correlated with missing description. It would indicate that either longer book or shorter books had a missing description and we are not going to use the books that have a missing description so it means that our recommender may be biassed towards shorter or longer books same as age of the book, average rating.

If we look at the values above we can see that they are really not very strong.

```python
books_missing = df[~(df["description"].isna())&
~(df["num_pages"].isna())&
~(df["average_rating"].isna()) &
~(df["published_year"].isna())
]
```
